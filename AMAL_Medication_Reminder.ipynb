```python
medications = []

def add_medication():
    name = input("Enter medication name: ")
    time = input("Enter time (HH:MM): ")
    medications.append((name, time))
    print(f"{name} at {time} added!")

def show_medications():
    print("\nCurrent Medications:")
    for med in medications:
        print(f"{med[0]} at {med[1]}")
    print("\n")

while True:
    print("1. Add Medication\n2. Show Medications\n3. Exit")
    choice = input("Choose an option: ")
    if choice == '1':
        add_medication()
    elif choice == '2':
        show_medications()
    elif choice == '3':
        break
    else:
        print("Invalid choice!")
```
